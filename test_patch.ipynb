# RF antenna: microstrip fed patch
We simulate a narrowband patch antenna fed by a microstrip line around 1.5 GHz. Compare with [COMSOL example](https://www.comsol.com/model/microstrip-patch-antenna-11742).

## Geometry

```python
import luminescent as lumi
from gdsfactory.technology import LogicalLayer, LayerLevel, LayerStack
import gdsfactory as gf
import matplotlib.pyplot as plt
import numpy as np
import os

# length and frequency units are arbitrary but must be consistent. Here we use mm and GHz.
frequency = 1.55  # characteristic frequency in GHz
frequencies = np.linspace(1.45, 1.65, 101).tolist()
wl1f = 300  # vacuum wavelength at 1 unit of frequency in your units
wavelength = wl1f / frequency  # characteristic wavelength

# microstrip patch [mm]
d = 1.524 # sub depth
w_line = 3.2 # 50 ohm line width
w_patch = 53 # Patch width
l_patch = 52 # Patch length
w_slot = 7 # slot width
l_flap = 15.5 # flap or slot length
w_flap=w_patch/2-w_slot-w_line/2
w_sub = 200 # sub width
l_sub = 200 # sub length
th=.1 # metal thickness

lateral_port_margin = height_port_margin = 2*d
source_port_margin =  1.5*(w_line+2*lateral_port_margin)
lx_mode = w_line + 2 * lateral_port_margin
ly_mode = th + height_port_margin + d

# layers
WG = 1, 0
PATCH = 2, 0
SUB= 3, 0

# make geometry in gdsfadory
# alternatively can import .gds into gdsfadory

c = gf.Component()

c.add_polygon(
    [
        (-l_patch / 2, -w_patch / 2),
        (-l_patch / 2 + l_flap, -w_patch / 2),
        (-l_patch / 2 + l_flap, -w_patch / 2 + w_flap),
        (-l_patch / 2, -w_patch / 2 + w_flap),
    ],
    layer=PATCH,
)
c.add_polygon(
    [
        (-l_patch / 2, w_patch / 2),
        (-l_patch / 2 + l_flap, w_patch / 2),
        (-l_patch / 2 + l_flap, w_patch / 2 - w_flap),
        (-l_patch / 2, w_patch / 2 - w_flap),
    ],
    layer=PATCH,
)
c.add_polygon(
    [
        (l_patch / 2, -w_patch / 2),
        (-l_patch / 2 + l_flap, -w_patch / 2),
        (-l_patch / 2 + l_flap, w_patch / 2),
        (l_patch / 2, w_patch / 2),
    ],
    layer=PATCH,
)

c.add_polygon(
    [
        (-l_sub / 2, -w_sub / 2),
        (-l_sub / 2, w_sub / 2),
        (l_sub / 2, w_sub / 2),
        (l_sub / 2, -w_sub / 2),
    ],
    layer=SUB,
)

line = c << gf.components.straight((l_sub - l_patch) / 2 + l_flap, width=w_line)
line.movex(-l_sub / 2)
c.add_port(
    f"o1", center=(-l_patch / 2, 0), width=w_line, orientation=180, layer=WG
)

c.plot()
```

## Solve

```python
# adaptive meshing
nres = 16 # points per wavelength in material
dx0 = wavelength / nres  # mesh step in air

layer_stack = LayerStack(
    layers={
        "patch": LayerLevel(
            layer=LogicalLayer(layer=PATCH),
            zmin=0,
            thickness=th,
            material="PEC",
            mesh_order=1,
        ),
        "line": LayerLevel(
            layer=LogicalLayer(layer=WG),
            zmin=0,
            thickness=th,
            material="PEC_line",
            mesh_order=2,
        ),
        "core": LayerLevel(
            layer=LogicalLayer(layer=SUB),
            zmin=-d,
            thickness=d,
            material="sub",
            mesh_order=3,
        ),
    }
)

dx_line = w_line / 6  # 6 mesh points across the strip
dx_patch = l_patch / 20  # 20 mesh points across the patch

eps_sub = 3.38
n_sub = np.sqrt(eps_sub)

material_library = {
    "air": lumi.Material(),
    "sub": lumi.Material(epsilon=eps_sub),
    "PEC": lumi.PECMaterial(mesh_density=dx0/dx_patch),
    "PEC_line": lumi.PECMaterial(mesh_density=[n_sub, dx0/dx_line,1]),
    'background': lumi.Material(),
}

# radiation port
R=H = (w_sub) / 2
angres=np.deg2rad(5)
max=np.deg2rad(80) # avoid microstrip
max=np.deg2rad(360)
ports = [
    lumi.SpherePort(
        name="o2",
        origin=(0, 0, 0),
        radius=R,
        angres=angres,
        max=max,
        max=max,
    ),
]

# microstrip mode
_dx = dx_line
_dy = th
modes = [
    lumi.Mode(
        ports=["o1"],
        frequencies=np.linspace(frequencies[1], frequencies[-1], 5),
        metallic_boundaries=["y-", "line"],
        nmodes=1,
        start=[-lx_mode / 2, -d,],  # local xy frame, local y=0 at global zmin of port layer
        stop=[lx_mode / 2, th + height_port_margin],
        voltage_line=[(0, -d), (0, 0)],  # optional
        current_loop=[
            (-w_line / 2 - _dx, -_dy),
            (-w_line / 2 - _dx, th + _dy),
            (w_line / 2 + _dx, th + _dy),
            (w_line / 2 + _dx, -_dy),
        ],  # optional
    )
]

sources = [
    lumi.Source(
        "o1", frequency=frequency, duration=1, source_port_margin=source_port_margin
    ) # modulated Gaussian pulse
]

material_color_map={
        'background':'transparent',
        'sub':('yellow',.5),
        'PEC':('black',.5), # (color, alpha)
}
views=[
        lumi.View("Ex", z=th / 2,  field_color_intensity=10,material_color_map=material_color_map,material_library=material_library),
        # lumi.View("Ez", z=-d / 2,  field_color_intensity=10,material_color_map=material_color_map,material_library=material_library),
    ]

# mounts Luminescent community cloud in local folder
cloud='luminescent_community_cloud'
mount=os.path.expanduser(f'~/{cloud}')
os.makedirs(mount, exist_ok=True)
os.system(f'gcsfuse --anonymous-access {cloud} {mount}')
email=____

# simulation folder in mounted cloud folder
study='microstrip_patch_antenna'
path = os.path.join(mount,email, study)

lumi.make(
    path=path,  # path to make problem folder
    component=c,
    frequencies=frequencies,
    frequency=frequency,
    wl1f=wl1f,
    boundaries=["PML", "PML", ["PEC", "PML"]],
    #
    ports=ports,
    modes=modes,
    sources=sources,
    #
    zmin=-d,
    zmax=H,
    #
    material_library=material_library,
    layer_stack=layer_stack,
    #
    nres=nres,  # number of grid points per wavelength in material (not vacuum)
    relative_courant=0.9,
    energy_decay_threshold=.001,  # stop when energy have decayed to this fraction of peak
    saveat=.5,  # save frame every _ periods for plotting movie
    # visualization
    views=views,
)
```

## Analysis

```python
sol = lumi.load(path)

# S11 plot
fig = plt.figure()
x = frequencies
y = [10 * np.log10(v) for v in lumi.query(sol, "T1,1")]
# y = [10 * math.log10(v) for v in lumi.query(sol, "To2@0,o1@0")]

plt.plot(x, y)
plt.xlabel("f [GHz]")
plt.ylabel("S11 [dB]")
plt.show()

flux = sol["dBi"]["o2"] # can append after o2 @H, V, L or R for polarization
i = np.argmin(y)
n = round(max / angres) + 1
n = round(max / angres) + 1
 = np.linspace(0, max, n)

# Radiation pattern plots
flux_E_plane1 = flux[:, 0, i]
flux_E_plane2 = flux[:, round((n - 1) / 2), i]

flux_H_plane1 = flux[:, round((n - 1) / 4), i]
flux_H_plane2 = flux[:, round((n - 1) / 4 * 3), i]

fig = plt.figure()

ax = fig.add_subplot(121, polar=True)
ax.set_theta_zero_location("N")
ax.set_theta_direction("clockwise")
ax.plot(, flux_E_plane1)
ax.plot(-, flux_E_plane2)
ax.set_title("E-Plane Radiation Pattern")
ax.set_ylim(-30, 30)

ax = fig.add_subplot(122, polar=True)
ax.set_theta_zero_location("N")
ax.set_theta_direction("clockwise")
ax.plot(, flux_H_plane1)
ax.plot(-, flux_H_plane2)
ax.set_ylim(-30, 30)
ax.set_title("H-Plane Radiation Pattern")
```

```python
y=Z0 = 377 * sol["Z"]["o1"]
plt.plot(x, np.real(y), label="Real", color="orange")
plt.plot(x, np.imag(y), label="Imaginary", color="blue")
plt.xlabel("f [GHz]")
plt.ylabel("complex line impedance [Ohms]")
plt.legend()
plt.show()
```
